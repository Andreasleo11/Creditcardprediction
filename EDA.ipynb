```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
merged_bureau['DAYS_CREDIT_ENDDATE'] = imputer.fit_transform(merged_bureau[['DAYS_CREDIT_ENDDATE']])
merged_bureau['DAYS_ENDDATE_FACT'] = imputer.fit_transform(merged_bureau[['DAYS_ENDDATE_FACT']])
merged_bureau['AMT_CREDIT_MAX_OVERDUE'] = imputer.fit_transform(merged_bureau[['AMT_CREDIT_MAX_OVERDUE']])
merged_bureau['AMT_CREDIT_SUM'] = imputer.fit_transform(merged_bureau[['AMT_CREDIT_SUM']])
merged_bureau['AMT_CREDIT_SUM_DEBT'] = imputer.fit_transform(merged_bureau[['AMT_CREDIT_SUM_DEBT']])
merged_bureau['AMT_CREDIT_SUM_LIMIT'] = imputer.fit_transform(merged_bureau[['AMT_CREDIT_SUM_LIMIT']])
merged_bureau['AMT_ANNUITY'] = imputer.fit_transform(merged_bureau[['AMT_ANNUITY']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Credit Active Status Distribution
credit_active_counts = merged_bureau['CREDIT_ACTIVE'].value_counts()
print("Credit Active Status Distribution:\n", credit_active_counts)

# Credit Currency Distribution
credit_currency_counts = merged_bureau['CREDIT_CURRENCY'].value_counts()
print("\nCredit Currency Distribution:\n", credit_currency_counts)

# Credit Type Distribution
credit_type_counts = merged_bureau['CREDIT_TYPE'].value_counts()
print("\nCredit Type Distribution:\n", credit_type_counts)

# Days of Credit Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=merged_bureau, x='DAYS_CREDIT', bins=50)
plt.xlabel('Days of Credit')
plt.ylabel('Frequency')
plt.title('Days of Credit Distribution')
plt.show()

# Credit Overdue and Prolongation Distribution
credit_overdue_counts = merged_bureau['CREDIT_DAY_OVERDUE'].value_counts()
credit_prolong_counts = merged_bureau['CNT_CREDIT_PROLONG'].value_counts()
print("\nCredit Overdue Distribution:\n", credit_overdue_counts)
print("\nCredit Prolongation Distribution:\n", credit_prolong_counts)

# Credit Sum and Debt Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=merged_bureau, x='AMT_CREDIT_SUM', bins=50, label='Credit Sum')
sns.histplot(data=merged_bureau, x='AMT_CREDIT_SUM_DEBT', bins=50, label='Credit Debt')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.title('Credit Sum and Debt Distribution')
plt.legend()
plt.show()

# Credit End Dates Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=merged_bureau, x='DAYS_CREDIT_ENDDATE', bins=50, label='Credit End Date')
sns.histplot(data=merged_bureau, x='DAYS_ENDDATE_FACT', bins=50, label='Credit End Date Fact')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.title('Credit End Dates Distribution')
plt.legend()
plt.show()

# Credit Max Overdue Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=merged_bureau, x='AMT_CREDIT_MAX_OVERDUE', bins=50)
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.title('Credit Max Overdue Distribution')
plt.show()

# Annuity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=merged_bureau, x='AMT_ANNUITY', bins=50)
plt.xlabel('Annuity Amount')
plt.ylabel('Frequency')
plt.title('Annuity Distribution')
plt.show()

# Monthly Balance and Status
monthly_balance_counts = merged_bureau['MONTHS_BALANCE'].value_counts()
status_counts = merged_bureau['STATUS'].value_counts()
print("\nMonthly Balance Distribution:\n", monthly_balance_counts)
print("\nStatus Distribution:\n", status_counts)

# Correlation Matrix
correlation_matrix = merged_bureau.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.impute import SimpleImputer

# Separate numerical and categorical columns
numerical_columns = previous_application.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = previous_application.select_dtypes(include=['object']).columns

# Impute numerical columns with mean
numerical_imputer = SimpleImputer(strategy='mean')
previous_application[numerical_columns] = numerical_imputer.fit_transform(previous_application[numerical_columns])

# Impute categorical columns with a special category ('MISSING')
categorical_imputer = SimpleImputer(strategy='constant', fill_value='MISSING')
previous_application[categorical_columns] = categorical_imputer.fit_transform(previous_application[categorical_columns])

# Print the first few rows of the cleaned DataFrame
print(previous_application.head())
```

```python
# Separate numerical and categorical columns
numerical_columns = pos_cash.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = pos_cash.select_dtypes(include=['object']).columns

# Impute numerical columns with mean
numerical_imputer = SimpleImputer(strategy='mean')
pos_cash[numerical_columns] = numerical_imputer.fit_transform(pos_cash[numerical_columns])

# Impute categorical columns with a special category ('MISSING')
categorical_imputer = SimpleImputer(strategy='constant', fill_value='MISSING')
pos_cash[categorical_columns] = categorical_imputer.fit_transform(pos_cash[categorical_columns])

# Print the first few rows of the cleaned DataFrame
print(pos_cash.head())
```
